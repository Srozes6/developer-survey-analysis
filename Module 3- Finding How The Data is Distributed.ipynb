### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
df.columns
df.dtypes
df.shape
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
df.isnull().sum()
df = df.ffill()
df.isnull().sum()

```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
df['Employment'].value_counts()
df['JobSat'].value_counts().sort_index()
df['YearsCodePro'].value_counts()
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
sns.kdeplot(df['JobSat'], fill=True)
plt.title('Job Satisfaction Distribution')
plt.xlabel('Job Satisfaction')
plt.show()


# The KDE plot illustrates how job satisfaction values are distributed across respondents. The concentration of values toward the higher end suggests that most respondents report positive job satisfaction, with fewer respondents reporting low satisfaction levels.
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
# Select relevant columns and drop missing values
lang_df = df[['LanguageHaveWorkedWith', 'LanguageWantToWorkWith']].dropna()

# Split language strings into lists
lang_df['LanguageHaveWorkedWith'] = lang_df['LanguageHaveWorkedWith'].str.split(';')
lang_df['LanguageWantToWorkWith'] = lang_df['LanguageWantToWorkWith'].str.split(';')

# Count languages worked with
have_counts = lang_df['LanguageHaveWorkedWith'].explode().value_counts()

# Count languages wanted to work with
want_counts = lang_df['LanguageWantToWorkWith'].explode().value_counts()

language_counts = pd.DataFrame({
    'Have Worked With': have_counts,
    'Want to Work With': want_counts
}).fillna(0)

language_counts = language_counts.head(10)

language_counts.plot(
    kind='bar',
    figsize=(10, 6)
)

plt.title('Programming Languages: Have Worked With vs Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# The grouped bar chart compares programming languages that respondents have worked with versus those they want to work with. Some languages show consistently high usage and interest, while others appear more frequently in the “want to work with” category, indicating emerging trends and future skill preferences among developers.
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
remote_df = df[['Country', 'RemoteWork']].dropna()
top_countries = remote_df['Country'].value_counts().head(5).index
remote_df = remote_df[remote_df['Country'].isin(top_countries)]

remote_ct = pd.crosstab(
    remote_df['Country'],
    remote_df['RemoteWork']
)

sns.heatmap(remote_ct, annot=True, fmt='d', cmap='Blues')
plt.title('Remote Work Distribution by Region')
plt.xlabel('Remote Work Preference')
plt.ylabel('Country')
plt.show()

# The visualization shows how remote work preferences differ across regions. Some regions report higher levels of remote or hybrid work, while others show a stronger preference for in-person work. This suggests that geographic location influences access to and adoption of remote work opportunities.
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
corr_df = df[['JobSat', 'YearsCodePro']].copy()

corr_df['YearsCodePro'] = pd.to_numeric(corr_df['YearsCodePro'], errors='coerce')
corr_df['JobSat'] = pd.to_numeric(corr_df['JobSat'], errors='coerce')

corr_df = corr_df.dropna()

pearson_corr = corr_df['JobSat'].corr(corr_df['YearsCodePro'], method='pearson')
pearson_corr

# The Pearson and Spearman correlation coefficients indicate the strength and direction of the relationship between professional coding experience and job satisfaction. The results suggest that job satisfaction does not strongly increase or decrease with years of experience, indicating that factors beyond experience alone may influence overall job satisfaction.
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
edu_emp_df = df[['Employment', 'EdLevel']].dropna()
edu_emp_ct = pd.crosstab(
    edu_emp_df['EdLevel'],
    edu_emp_df['Employment']
)

edu_emp_ct
edu_emp_norm = pd.crosstab(
    edu_emp_df['EdLevel'],
    edu_emp_df['Employment'],
    normalize='index'
)

edu_emp_norm
edu_emp_norm.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6)
)

plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion of Respondents')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# The stacked bar chart illustrates how employment status varies across education levels. Higher levels of education show a greater proportion of full-time employment, while lower education levels display more variation across employment types. This suggests that educational attainment is associated with employment outcomes.
```
